<span style="color:magenta;font-size:16px;">

<strong>What is IoT?</strong><br>
Internet of Things (IoT) refers to devices embedded with sensors and actuators that communicate with computing systems via wired or wireless networks—allowing the physical world to be digitally monitored or even controlled. Things we use in day to day life such as fitness trackers are considered part of IoT.  IoT uses a variety of technologies to connect the digital and physical worlds. Physical objects are embedded with sensors—which can monitor things like temperature or motion, or really any change in environment—and actuators—which receive signals from sensors and then do something in response to those changes. Since IoT devices are connected to the internet, their vulnerability to cyber-attack is very high. There are various attacks, each of which are discussed below.
</span>

<span style="color:magenta;font-size:16px;">

<strong>Distributed Denial-of-Service (DDoS)</strong><br>
DDoS attacks aim to overwhelm a target system with a massive surge of traffic, rendering it inaccessible to legitimate users. Attackers often employ botnets, networks of compromised devices, to amplify the attack's impact. In the IoT realm, DDoS attacks can target routers, smart home devices, or industrial control systems, causing disruptions in vital services.
</span>

<div>

<span style="color:magenta;font-size:16px;">

<strong>Brute Force</strong><br>
Brute force attacks involve systematically attempting various password combinations until the correct one is found. Attackers often rely on automated tools to rapidly test millions of possibilities, particularly targeting devices with weak or default passwords. Once compromised, IoT devices can be used for malicious purposes like data theft, unauthorized access, or participation in coordinated attacks.
</span>

<div>

<span style="color:magenta;font-size:16px;">

<strong>Spoofing</strong><br>
Spoofing attacks involve masquerading as a trusted entity to deceive devices or networks. Attackers may forge IP or MAC addresses to gain unauthorized access, intercept or manipulate data exchanges, or compromise the integrity of communication channels. Spoofing poses a significant threat to IoT systems that rely on mutual trust between devices for secure operations.
</span>

<span style="color:magenta;font-size:16px;">

<strong>Denial-of-Service (DoS)</strong><br>
DoS attacks, like DDoS attacks, aim to disrupt the availability of a service or network. However, DoS attacks typically originate from a single source. Attackers may overwhelm the target with traffic or exploit vulnerabilities to crash the system, rendering it temporarily or permanently unavailable. In the IoT context, DoS attacks can disrupt the normal functioning of devices, preventing them from providing critical services, especially in applications where real-time communication is essential.

</span>

<span style="color:magenta;font-size:16px;">

<strong>Reconnaissance</strong><br>
Reconnaissance attacks involve gathering information about a target system to identify vulnerabilities and potential entry points. Attackers employ various methods, such as network scanning, probing for weaknesses, and collecting information about device configurations. In the IoT domain, reconnaissance attacks may focus on identifying poorly secured devices or exploiting vulnerabilities in IoT communication protocols. Successful reconnaissance sets the stage for more targeted and damaging attacks.
</span>

<span style="color:magenta;font-size:16px;">

<strong>Web-based</strong><br>
Web-based attacks exploit vulnerabilities in web applications or services. Attackers target weaknesses in web servers, databases, or application code to gain unauthorized access, manipulate data, or compromise user privacy. In the IoT context, web-based attacks may focus on exploiting vulnerabilities in web interfaces used for managing IoT devices. Compromising these interfaces can allow attackers to control or manipulate connected devices, posing various security and privacy risks.

Mirai is a type of malware specifically designed to infect and control IoT devices, forming a botnet. Compromised devices can be used to launch large-scale DDoS attacks. Mirai targets devices with weak security, often using default usernames and passwords. The goal is to create a network of compromised devices that can be centrally controlled to execute coordinated attacks.

2.5. PCA. Principal Component Analysis (PCA) is a powerful technique for dimensionality reduction and
visualization of high-dimensional data. Applying PCA to the dataset can help visualize the separation
(or lack thereof) between different attack groups in a 2D or 3D space. After applying PCA to the
numerical columns, scatter plot the first two or three principal components. Hint: PCA is sensitive to
the magnitude of the data, so you should standardize the features to have mean=0 and variance=1.

<span style="color:magenta;font-size:16px;">

Doing PCA can help to reduce the dimension of our dataset. First, we will retrieve the numerical columns.

</span>

````python
num_cols = combined_df.select_dtypes(include=[np.number]) #get numeric cols
num_cols
````

<span style="color:magenta;font-size:16px;">

Next we scale the data and do PCA on our scaled data.

</span>

````python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
normalized_c = scaler.fit_transform(num_cols)
normalized_c
normalized = pd.DataFrame(data=normalized_c, columns=num_cols.columns)
````

<span style="color:magenta;font-size:16px;">

In order to do PCA we need the covariance matrix, eigen values, and eigen vectors.

</span>

````python
covariance_matrix = np.cov(normalized.values.T)
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
idx = eigen_values.argsort()[::-1]
eigen_values = eigen_values[idx]
eigen_vectors = eigen_vectors[idx]
eigen_vectors
````

<span style="color:magenta;font-size:16px;">

The following code computes the variance explained and the cumulative so that we can easily identify which components have the major role in the variance of our data.

</span>

````python
variance_explained = []
for i in eigen_values:
    variance_explained.append(i/np.sum(eigen_values)*100)
cumulative_ve = np.cumsum(variance_explained)
cumulative_ve
````

````python
selected_indices = np.where(eigen_values > 1)
selected_eigenvalues = eigen_values[selected_indices]
selected_eigenvectors = eigen_vectors[:, selected_indices]
selected_eigenvalues
````

<span style="color:magenta;font-size:16px;">

The following code visualizes where does our cumulative reach 95% mark for the cutoff. 

</span>

````python
plt.plot(np.arange(1,len(cumulative_ve)+1,1), cumulative_ve, marker = '*', linewidth=2, color='b', markersize=12, label='Cummulative Sum');
plt.plot(np.arange(1,len(cumulative_ve)+1,1), (np.ones((1,len(cumulative_ve)))*95).flatten(), color = 'r', linewidth=2, markersize=10, label='95% cutoff');
plt.grid()
plt.xlabel('Number of principal components.',fontsize=12);
plt.ylabel('Cummulative variance explanation %.', fontsize=12);
plt.title('Cummulative Variance Explanation vs Number of Principal Components.',fontsize=12);
plt.xticks(fontsize=12);
plt.yticks(fontsize=12);
plt.legend(fontsize=12);
````
